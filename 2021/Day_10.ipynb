```python
with open("input_10.txt", "r") as input_file:
    values = [val.rstrip() for val in input_file.readlines()]

opens = ['(','[','{','<']

def score(char):
    scores = {
        ')': 3,
        ']': 57,
        '}': 1197,
        '>': 25137
    }
    return scores.get(char, 0)
def closes(char):
    matchers = {
        ')': '(',
        ']': '[',
        '}': '{',
        '>': '<'
    }
    return matchers.get(char, 'invalid')

totalScore = 0
InvalidRows = []

for index, row in enumerate(values):
    stack = []
    for char in row:
        if(char in opens):
            stack.append(char)
        else:
            if(stack[-1] == closes(char)):
                stack.pop()
            else:
                totalScore += score(char)
                InvalidRows.append(index)
                break



print('Task 1:', totalScore)


# Part 2 (x-1) if (x != 0) else 6 
incompletes = list(filter(lambda a: a != '',[row if (index not in InvalidRows) else '' for index,row in enumerate(values)]))

```
