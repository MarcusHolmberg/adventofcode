```python
import numpy

# Task 1
with open("input_4.txt", "r") as input_file:
    inputlines = [line.rstrip() for line in input_file]

# numbers example ['1','2','3']
numbers = inputlines[0].split(',')

# bingogrid sample:
# [['1','2','3'],
#  ['4','5','6'],
#  ['7','8','9']]
bingogridsbyrows = []
bingogridsbycolumns = []
tempgrid = []
for index, line in enumerate(inputlines[2:]):
    if(line != ''):
        gridline = line.rstrip().split()
        tempgrid.append(gridline)
        if(len(tempgrid) == 5):
            bingogridsbyrows.append(tempgrid)
            tempgrid = []
        

for grid in bingogridsbyrows:
    bingogridsbycolumns.append(numpy.transpose(grid))

def validRow(nums, row):
    return all(elem in nums for elem in row)

def uncalledSum(callednums, grid):
    uncalled = []
    for row in grid:
        for num in row:
            if(num not in callednums):
                uncalled.append(int(num))
    print(uncalled)
    return sum(uncalled)

def findWinner(gridCollection):
    callednums = []
    for num in numbers:
        callednums.append(num)
        for index, grid in enumerate(gridCollection):
            for row in grid:
                if(validRow(callednums, row)):
                    return index,num,grid,callednums



r_index,r_num,r_grid,r_callednums = findWinner(bingogridsbyrows)                
c_index,c_num,c_grid,c_callednums = findWinner(bingogridsbycolumns)
#print(r_index,r_num,r_grid,r_callednums)
#print(c_index,c_num,c_grid,c_callednums)
print('Task 1:',uncalledSum(c_callednums, c_grid),'*',c_num,'=',int(c_num)*(uncalledSum(c_callednums,c_grid)))

# Task 2
def findLastWinner(gridCollection):
    callednums = []
    winners = set()
    for num in numbers:
        callednums.append(num)
        for index, grid in enumerate(gridCollection):
            for row in grid:
                if(validRow(callednums, row)):
                    winners.add(index)
                    if(len(winners) == len(gridCollection)):
                        print(len(winners),index,num)
                        return index,num,grid,callednums


r2_index,r2_num,r2_grid,r2_callednums = findLastWinner(bingogridsbyrows)                
c2_index,c2_num,c2_grid,c2_callednums = findLastWinner(bingogridsbycolumns)

#print(c2_callednums)
#print(bingogridsbyrows[c2_index])
#print(r2_index,r2_num,r2_grid,r2_callednums)
#print(c2_index,c2_num,c2_grid,c2_callednums)

print('Task 2:',uncalledSum(c2_callednums,bingogridsbyrows[c2_index]),'*',c2_num,'=',int(c2_num)*(uncalledSum(c2_callednums,bingogridsbyrows[c2_index])))
```
