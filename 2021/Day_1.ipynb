```python
# Task 1
with open("input_1.txt", "r") as input_file:
    values = [int(line.rstrip()) for line in input_file]

count = 0
last = 148
for current in values:
    if current > last:
        count = count + 1
    last = current

print(count)
```

```python
# Task 2
with open("input_1.txt", "r") as input_file:
    values = [int(line.rstrip()) for line in input_file]

summed = []

for index, current in enumerate(values):
    if(index < len(values)-2):
        sum = current + values[index+1] + values[index+2]
        summed.append(sum)

count = 0
last = 483
for current in summed:
    if current > last:
        count = count + 1
    last = current

print(count)
```
