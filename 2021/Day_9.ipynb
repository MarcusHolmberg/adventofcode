```python
import numpy as np
from scipy import spatial

with open("input_9.txt", "r") as input_file:
    values = [val.rstrip() for val in input_file.readlines()]
rows = []
for row in values:
    rows.append([int(x) for x in list(row)])

numpyArr = np.array(rows)

lowspots = 0
lowspotsPos = []

for i1,row in enumerate(rows):
    for i2,num in enumerate(row):
        isLow = True
        if(i1 > 0):
            #Check up
            if(rows[i1-1][i2] <= num):
                isLow = False
        if(i1 < len(rows)-1):
            #Check down
            if(rows[i1+1][i2] <= num):
                isLow = False
        if(i2 > 0):
            #check left
            if(row[i2-1] <= num):
                isLow = False
        if(i2 < len(row)-1):
            #check right
            if(row[i2+1] <= num):
                isLow = False
        if(isLow):
            lowspots += num
            lowspotsPos.append([i1,i2])

sumOfLowSpotHieghts = lowspots + (len(lowspotsPos))

print('Task 1: ',sumOfLowSpotHieghts)

#468

#Task 2
#for every low spot, create a size, increase size for every number in same basin
#basin are contained by 9 and border
basins = [[x] for x in lowspotsPos]

basinPos = []
borderPos = []
xpos,ypos = np.where(numpyArr != 9)
bxpos, bypos = np.where(numpyArr == 9)



for i, x in enumerate(xpos):
    basinPos.append([xpos[i],ypos[i]])

for i, x in enumerate(bxpos):
    borderPos.append([bxpos[i],bypos[i]])


# for every number, find the nearest basin not blocked by borders

for i1,row in enumerate(rows):
    for i2,num in enumerate(row):
        if([i1,i2] in basinPos):
            if([i1,i2] not in lowspotsPos):
              print('Nearest lowspot position: ',lowspotsPos[spatial.KDTree(lowspotsPos).query([i1,i2])[1]])
```
